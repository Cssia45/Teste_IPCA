```
from pyspark.sql import SparkSession

# Inicializa a SparkSession
spark = SparkSession.builder \
.appName("Export JSON to Delta Lake") \
.getOrCreate()

# Caminho para o arquivo JSON
json_file_path = "/FileStore/tables/dados.json"

# Lê o arquivo JSON
df = spark.read.json(json_file_path)

# Cria a tabela Dados_IPCA em formato delta lake a partir do arquivo json
df.write.format("delta").saveAsTable("Dados_IPCA")

```
